```python
import numpy as np
import pandas as pd

s = ['희빈','수영','현수','지호','지민']
p = pd.Series([99,84,92,65,78],index=s)
p
```

```python
p.name = '수학성적'
p.index.name = '이름'
p
```

```python
p.sort_values()
```

```python
p.sort_values(ascending=False)
```

```python
p.sort_index()
```

```python
p.sort_index(ascending=False)
```

```python
print(p.max())
print(p.min())
print(p.mean())
print(p.median())
print(p.count())
print(p.size)
```

```python
s = pd.Series([3,1,1,2, None,np.nan,4,6,6,np.nan])
s
```

```python
s.values
```

```python
s.count() #nan은 카운트에서 제외
```

```python
s.value_counts() #가장 많은거부터 카운드를 세준다.
```

```python
# s.value_counts(dropna=False)
s.value_counts(dropna=True)
```

```python
s.value_counts(sort=False,dropna=False)
```

```python
s.value_counts(bins=4)
```

```python
s.value_counts(bins=4,sort=False)
```

```python
s.value_counts(bins=[0.5,3.5,6.5],sort=False)
```
